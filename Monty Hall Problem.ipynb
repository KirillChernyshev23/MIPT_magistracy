```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
iterations = 100000 #общее кол-во игр

#переменные для варианта без смены двери
car_count1 = 0
goat_count1 = 0
num_iter1 = np.zeros(iterations)
win_prob_iter1 = np.zeros(iterations)

#переменные для "умного:)" варианта
car_count2 = 0
goat_count2 = 0
num_iter2 = np.zeros(iterations)
win_prob_iter2 = np.zeros(iterations)
```

```python
#основной цикл

for i in range(iterations):
    choices = ["Door", "Door", "Door"]
    choices[random.randint(0, 2)] = "Car"
    if choices[0] == "Car":
        car_count1 += 1
    num_iter1[i] = i + 1    
    win_prob_iter1[i] = float(car_count1/num_iter1[i]) 
    if (choices[0] == "Door"):
        car_count2 += 1
    num_iter2[i] = i + 1    
    win_prob_iter2[i] = float(car_count2/num_iter2[i])
```

```python
#построение графиков

plt.plot(range(1, 100), win_prob_iter1[1: 100], label='without changing')
plt.plot(range(1, 100), win_prob_iter2[1: 100], label='with changing')
plt.xlabel('Count of Iterations')
plt.ylabel('p(n)')
plt.legend()
plt.show()
```

```python
plt.plot(range(1, 1000), win_prob_iter1[1: 1000], label='without changing')
plt.plot(range(1, 1000), win_prob_iter2[1: 1000], label='with changing')
plt.xlabel('Count of Iterations')
plt.ylabel('p(n)')
plt.legend()
plt.show()
```

```python
plt.plot(range(1, iterations + 1), win_prob_iter1, label='without changing')
plt.plot(range(1, iterations + 1), win_prob_iter2, label='with changing')
plt.xlabel('Count of Iterations')
plt.ylabel('p(n)')
plt.legend()
plt.show()
```

```python
print('При варианте без изменения изначального выбора p = ', win_prob_iter1[iterations - 1] )
print('При варианте c изменением изначального выбора p = ', win_prob_iter2[iterations - 1] )
```

Аналогично для 100 дверей.

```python
iterations = 100000 #общее кол-во игр

#переменные для варианта без смены двери
car_count1 = 0
goat_count1 = 0
num_iter1 = np.zeros(iterations)
win_prob_iter1 = np.zeros(iterations)

#переменные для "умного:)" варианта
car_count2 = 0
goat_count2 = 0
num_iter2 = np.zeros(iterations)
win_prob_iter2 = np.zeros(iterations)
```

```python
for i in range(iterations):
    choices = ["Door"] * 100
    choices[random.randint(0, 99)] = "Car"
    if choices[0] == "Car":
        car_count1 += 1
    num_iter1[i] = i + 1    
    win_prob_iter1[i] = float(car_count1/num_iter1[i]) 
    if (choices[0] == "Door"):
        car_count2 += 1
    num_iter2[i] = i + 1    
    win_prob_iter2[i] = float(car_count2/num_iter2[i])
```

```python
plt.plot(range(1, 100), win_prob_iter1[1: 100], label='without changing')
plt.plot(range(1, 100), win_prob_iter2[1: 100], label='with changing')
plt.xlabel('Count of Iterations')
plt.ylabel('p(n)')
plt.legend()
plt.show()
```

```python
plt.plot(range(1, iterations + 1), win_prob_iter1, label='without changing')
plt.plot(range(1, iterations + 1), win_prob_iter2, label='with changing')
plt.xlabel('Count of Iterations')
plt.ylabel('p(n)')
plt.legend()
plt.show()
```

```python
print('При варианте без изменения изначального выбора (при количестве дверей, равным 100) p = ', win_prob_iter1[iterations - 1] )
print('При варианте c изменением изначального выбора(при количестве дверей, равным 100) p = ', win_prob_iter2[iterations - 1] )
```
